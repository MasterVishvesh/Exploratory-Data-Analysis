```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
print(iris.shape)
# print(iris.head)
```

```python
print(iris.columns)
print(iris["species"].value_counts())
iris.plot(kind='scatter', x='sepal_length', y='sepal_width');
plt.show()
```

```python
sns.set_style("darkgrid");
sns.FacetGrid(iris, hue="species", height=5)\
   .map(plt.scatter, "sepal_length", "sepal_width")\
   .add_legend();
plt.show();
```

```python
#pair-polt
#can be use if number of features are high
plt.close();
sns.set_style("darkgrid");
sns.pairplot(iris, hue="species", size=3);
plt.show();
```

```python
iris_setosa = iris.loc[iris["species"] == "setose"];
iris_versicolor = iris.loc[iris["species"] == "versicolor"];
iris_virginica = iris.loc[iris["species"] == "virginica"];
# print(iris_setosa["petal_length"])
plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa["petal_length"]), "o", color="red");
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor["petal_length"]), "o", color="blue");
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica["petal_length"]), "o", color="green");
plt.show()
```

```python
sns.FacetGrid(iris, hue="species", height=5) \
   .map(sns.distplot, "petal_length") \
   .add_legend();
plt.show();
```

```python
#univararite analysis

# cdf and pda

counts, bin_edges = np.histogram(iris_setosa["petal_length"], bins=10, density=True);

pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);

#compute CDF
cdf = np.cumsum(pdf);
plt.plot(bin_edges[1:], cdf);
plt.show();
```

```python
# MEAN, ariance and std-dev

print(np.mean(iris_setosa["petal_length"]));
print(np.var(iris_setosa["petal_length"]));
print(np.std(iris_setosa["petal_length"]));
```
